```python
import urllib.request as ur
import urllib.parse as up
import urllib
#import lxml.html
import bs4 as bs
import os
from multiprocessing.pool import ThreadPool
```

```python
bases = ['https://www.psp.cz/eknih/2013ps/audio/2013/', 'https://www.psp.cz/eknih/2017ps/audio/2017/']
```

```python
bases_ht = [bs.BeautifulSoup(ur.urlopen(b), 'lxml') for b in bases]
```

```python
all_bases = bases.copy()
for base_year, base_url in zip(bases_ht, bases):
    other_years = base_year.cssselect('.simple a')
    for o_year in other_years:
        y_href = o_year.attrib['href']
        all_bases.append(up.urljoin(base_url, y_href))

all_bases = sorted(all_bases)
print(all_bases)
```

```python
bs_all_bases = [bs.BeautifulSoup(ur.urlopen(base), 'lxml') for base in all_bases]
```

```python
all_days = []
for soup_year, base_url in zip(bs_all_bases, all_bases):
    for day in soup_year.select('.event a'):
        all_days.append(ur.urljoin(base_url, day.get('href')))

print(all_days)
```

```python
bs_all_days = [bs.BeautifulSoup(ur.urlopen(day_url), 'lxml') for day_url in all_days]
```

```python
audio_urls = []
steno_urls = []

for bs_day, day_url in zip(bs_all_days, all_days):
    for link in bs_day.select('#main-content a'):
        link = link.get('href')
        if(link.endswith('.mp3')):
            audio_urls.append(ur.urljoin(day_url, link))
        else:
            steno_urls.append(ur.urljoin(day_url, link))
            
print(audio_urls)
print(len(audio_urls))
```

```python
audio_dir = '../shared/cpm_mp3/'
if not os.path.exists(audio_dir): os.mkdir(audio_dir)

def download_audio(a_url, folder=audio_dir):
    target = os.path.join(audio_dir, a_url.split('/')[-1])
    if os.path.isfile(target): return
    try:
        print('Retrieving ' + str(len(os.listdir(folder))) + '/' + str(len(audio_urls)) + ' ' + a_url + ' ', end='')
        ur.urlretrieve(a_url, target)
        print('done.')
    except urllib.error.HTTPError as err:
        print('failed: {}'.format(err), 'URL: ' + a_url)
```

```python
ThreadPool(12).imap_unordered(download_audio, audio_urls)
```

```python
def remove_small_audio(folder=audio_dir, limit=100*1024):
    for f in os.listdir(folder):
        full_f = os.path.join(folder, f)
        if os.path.getsize(full_f) < limit and os.path.isfile(full_f):
            os.remove(full_f)
```
